# Medical Data Visualizer

***

## Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

***

## Step by Step Process

<font color=blue>__Import data:__</font>

```python
df = pd.read_csv('medical_examination.csv')
df
```

<font color=blue>__Add 'overweight' column:__</font>

```python
df['overweight'] = (df['weight'] / ((df['height'] / 100) ** 2)).apply(lambda value: 1 if value > 25 else 0)

df
```

<font color=blue>__Normalize data by making 0 always good and 1 always bad:__</font>

```python
df['cholesterol'] = df['cholesterol'].apply(lambda value: 0 if value == 1 else 1)
df['gluc'] = df['gluc'].apply(lambda value: 0 if value == 1 else 1)

df
```

<font color=blue>__Draw Categorical Plot:__</font>

__Create DataFrame for cat plot using 'pd.melt', using just the values from 'cholesterol', 'gluc', 'smoke', 'alco', 'active', and 'overweight':__

```python
df_cat = ['cholesterol', 'gluc', 'smoke', 'alco', 'active', 'overweight']
df_cat
```

```python
df_cat = pd.melt(df, id_vars=['cardio'], value_vars=df_cat)
df_cat
```

__Group and reformat the data to split it by 'cardio'. Show the counts of each feature:__

```python
df_cat['count'] = 1
df_cat
```

```python
df_cat = df_cat.groupby(by=['cardio', 'variable', 'value'], as_index=False).count()
df_cat
```

__Draw the catplot with 'sns.catplot()':__

```python
fig = sns.catplot(x='variable', y='count', hue='value', kind='bar', col='cardio', data=df_cat).set_axis_labels('variable', 'total').fig
```

__Save the Cat Plot graphic:__

```python
fig.savefig('catplot.png')
```

<font color=blue>__Draw Heat Map:__</font>

```python
df_heat = df[
        (df['ap_lo'] <= df['ap_hi']) &
        (df['height'] >= df['height'].quantile(0.025)) &
        (df['height'] <= df['height'].quantile(0.975)) &
        (df['weight'] >= df['weight'].quantile(0.025)) &
        (df['weight'] <= df['weight'].quantile(0.975))
]

df_heat
```

__Calculate the correlation matrix:__

```python
corr = df_heat.corr()
corr
```

__Generate a mask for the upper triangle:__

```python
mask = np.zeros_like(corr)
mask
```

```python
mask[np.triu_indices_from(mask)] = True
mask
```

__Set up the matplotlib figure:__

```python
fig, ax = plt.subplots(figsize=(10, 10))
```

__Draw the heatmap with 'sns.heatmap()':__

```python
ax = sns.heatmap(corr, center=0.1, annot=True, fmt='.1f', linewidths=1, cbar_kws={'shrink': 0.7}, square=True, mask=mask)
ax
```

__Save the Heatmap graphic:__

```python
fig.savefig('heatmap.png')
```
